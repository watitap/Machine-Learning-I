### Import functions

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

### Load the dataset from the repository 

```python
data = pd.read_csv('wine.data')
data.to_csv('wine.data', header=['Class', 'Alcohol','Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline'], index=False)
```

### How disperse are the values in the dataset? 

```python
print('The Standard Deviation of each feature is shown below.\n', data.std(), '\n\n')
print('The Mean of each feature is shown below.\n', data.mean(), '\n\n')
print('Base on the data shown above, data is clustered in most feautures except Magnesium and Proline.')
```

### Create a scatter plot of two of the features of your choice. 

```python
plt.figure(figsize=(8,5))
newDataX = data['Alcohol']
newDataY = data['Proline']
plt.scatter(newDataX, newDataY, s=newDataX, c=newDataY, alpha=0.7)
plt.xlabel('Alcohol')
plt.ylabel('Proline')
plt.show()
```
