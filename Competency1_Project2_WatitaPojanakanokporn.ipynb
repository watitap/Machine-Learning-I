### Create a scatter plot of two of the features of your choice. 

```python
plt.figure(figsize=(10, 6))
newDataX = data['Alcohol']
newDataY = data['Proline']
plt.scatter(newDataX, newDataY, s=newDataX, c=newDataY, alpha=0.7)
plt.xlabel('Alcohol')
plt.ylabel('Proline')
plt.show()
```

Data is scattered but there is some positive correlation between Alcohol and Proline.

### Create a scatter matrix of at least 5 features of the dataset. 

```python
# Display Scatter Matrix
newFeatures = ['Alcohol', 'Proline', 'Hue', 'Ash', 'Malic acid', 'Color intensity']
scatter_matrix(data[newFeatures],figsize=(12, 12))
plt.show()
```

Data seems to be very scattered in all in the matrix, however, there are some positive correlations between Alcohol and Proline, some positive correlation between Alcohol and Color Intensity, and Proline and Color Intensity.

### Create any other plot of your preference. 

```python
# Display boxplot and outliers
plt.figure(figsize=(10,6))
boxplot = data.boxplot(column=['Alcohol', 'Hue', 'Ash', 'Malic acid', 'Color intensity'])
plt.show()
```

There is no outlier for Alcohol but there are outliers for Hue, Ash, Malic acid, and Color Intensity

```python
# Get data summary for the selected features
data[['Alcohol', 'Hue', 'Ash', 'Malic acid', 'Color intensity']].describe()
```

```python
# Get minimum, mean and maximum values to check where outliers are at
data_min = data[['Class','Alcohol', 'Hue', 'Ash', 'Malic acid', 'Color intensity']].min()
data_mean = data[['Class', 'Alcohol', 'Hue', 'Ash', 'Malic acid', 'Color intensity']].groupby('Class').mean()
data_max = data[['Class', 'Alcohol', 'Hue', 'Ash', 'Malic acid', 'Color intensity']].max()
```

```python
# Print mnimun, mean, and maximuv values for easy reading.
print('Minimum values:\n')
print(data_min, '\n\n')
print('Mean values:\n')
print(data_mean, '\n\n')
print('Maximum values:\n')
print(data_max)
```
