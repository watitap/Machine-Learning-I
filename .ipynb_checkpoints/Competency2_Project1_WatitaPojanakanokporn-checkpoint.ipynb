### Dataset information

Data represents Twelve posts' performance metrics extracted from a cosmetic company's page including 500 publications. We are finding what type of post works best for Facebook online advertising based on total interactions.

Reference: https://www.kaggle.com/datasets/mrsohelrana/facebook-metrics

### Import functions

```python
#%matplotlib inline

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

### Load datset and display first 6 rows with features

```python
url = 'https://www.kaggle.com/datasets/mrsohelrana/facebook-metrics'
data = pd.read_csv('dataset_Facebook.csv')
data.head(10)
```

### Data summary

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
data.describe()
```

### Describe data type of each feature

```python
data.info()
```

### Comments on data set

1. All except one feature are numeric type. "Type" column is string type. 
2. There are 500 instances(rows) and 19 features(columns)
3. The summary shows that Paid, like and share columns are some missing values.

### Check Paid, like and share columes for missing values

```python
paid_missing = data['Paid'].isna()
data.loc[paid_missing,:]
```

1 missing value is on row 499 of "Paid" column.

```python
like_missing = data['like'].isna()
data.loc[like_missing,:]
```

1 missing value are on row 111 of "like" column.

```python
share_missing = data['share'].isna()
data.loc[share_missing,:]
```

4 missing values are on row 111, 120, 124, and 164 of "share" column.

### Replace NaN with median value

```python
median_Paid = data['Paid'].median()
median_Paid
                
```

```python
median_like = data['like'].median()
median_like
```

```python
median_share = data['share'].median()
median_share
```

```python
### Fill Nan with median value
```

```python
data['Paid'].fillna(median_Paid, inplace=True)
data['like'].fillna(median_like, inplace=True)
data['share'].fillna(median_share, inplace=True)
```

```python
data.describe()
```

### Use boxplot to check outlier on likes

```python
plt.figure(figsize=(12, 5))
sns.boxplot(x ='Type', y='Total Interactions', data=data)
plt.xlabel('Type of Post')
plt.ylabel('Total Interactions')

plt.show()
```

### Create a scatter matrix of comment, like, share, and Total Interactions

```python
scatterData = ['comment','like', 'share', 'Total Interactions']
scatter_matrix(data[scatterData],figsize=(12, 12))
plt.show()
```

### Findings

1. There are outliers on Photo, Status and Link based on the result shown in scatter matrix.
2. There is positive corelation between like and Total Interactions based on the result shown in scatter matrix.
3. Video performs best based on Total interactions shown in boxplot.

### What challenges did you encounter when performing data exploration on this dataset

1. Missing values (NaN) need to be replaced and fill in
2. There is outlier in the dataset, we are not sure if it is good to keep or remove. At this time, we will keep the outlier
3. We were unsure whether to remove rows of missing values or replace missing values with some values such as zero, mean or median. We decided to reaplce with median value at this time.
