### Create a dataframe that includes only the features 

```python
data.plot(figsize=(8,10))
plt.xlabel('Numbers of data')
plt.ylabel('Amount')
plt.show()
```

### Create a scatter plot of two of the features of your choice. 

```python
plt.figure(figsize=(8,5))
newDataX = data['Alcohol']
newDataY = data['Proline']
plt.scatter(newDataX, newDataY, s=newDataX, c=newDataY, alpha=0.7)
plt.xlabel('Alcohol')
plt.ylabel('Proline')
plt.show()
```

### Create a scatter matrix of at least 5 features of the dataset. 

```python
newDataZ = ['Alcohol', 'Proline', 'Hue', 'Ash', 'Malic acid', 'Color intensity']
scatter_matrix(data[newDataZ],figsize=(8,6))
plt.show()
```
